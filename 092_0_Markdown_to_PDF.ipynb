```python
# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
# ğŸ“¦ å¿…è¦ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
!pip install weasyprint markdown

# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
# ğŸ“‚ ãƒ•ã‚¡ã‚¤ãƒ«ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰
# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
from google.colab import files
from IPython.display import display, FileLink

print("ğŸ”¼ Markdownãƒ•ã‚¡ã‚¤ãƒ«ã‚’ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã—ã¦ãã ã•ã„ï¼ˆ.mdï¼‰")
uploaded = files.upload()

# ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã•ã‚ŒãŸãƒ•ã‚¡ã‚¤ãƒ«åã‚’å–å¾—
filename = next(iter(uploaded))
with open(filename, 'r', encoding='utf-8') as f:
    markdown_text = f.read()

# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
# ğŸ” Markdown â†’ HTMLå¤‰æ›
# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
import markdown
html_body = markdown.markdown(markdown_text, extensions=['fenced_code', 'codehilite'])

# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
# ğŸ¨ ãƒãƒ¼ãƒˆé¢¨CSSï¼ˆã‚¹ã‚¿ã‚¤ãƒ«å®šç¾©ï¼‰
# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
notebook_css = """
body {
  font-family: "Courier New", monospace;
  background: #fffef8;
  color: #333;
  margin: 2em;
  line-height: 1.6;
}

h1, h2, h3 {
  border-bottom: 1px solid #ccc;
  padding-bottom: 0.2em;
}

code, pre {
  background: #f4f4f4;
  border: 1px solid #ddd;
  padding: 0.2em 0.4em;
  border-radius: 4px;
}

hr {
  border: none;
  border-top: 1px dashed #aaa;
  margin: 2em 0;
}
"""

# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
# ğŸ§¾ HTMLæ§‹ç¯‰ + PDFç”Ÿæˆ
# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
from weasyprint import HTML

html_full = f"""
<html>
<head>
<meta charset="utf-8">
<style>{notebook_css}</style>
</head>
<body>{html_body}</body>
</html>
"""

output_pdf = filename.replace(".md", "_notebook_style.pdf")
HTML(string=html_full).write_pdf(output_pdf)

# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
# ğŸ“¥ ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ãƒªãƒ³ã‚¯è¡¨ç¤º
# â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
print("âœ… PDFå¤‰æ›ãŒå®Œäº†ã—ã¾ã—ãŸã€‚ä»¥ä¸‹ã®ãƒªãƒ³ã‚¯ã‹ã‚‰ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã§ãã¾ã™ï¼š")
display(FileLink(output_pdf))
```
