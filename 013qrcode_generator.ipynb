```python
# @title QRã‚³ãƒ¼ãƒ‰ç”Ÿæˆãƒ„ãƒ¼ãƒ«

# å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
!pip install -q qrcode[pil] ipywidgets

# ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import qrcode
from PIL import Image
from IPython.display import display, clear_output
import ipywidgets as widgets
from google.colab import files

# å…¥åŠ›ã‚¦ã‚£ã‚¸ã‚§ãƒƒãƒˆï¼ˆãƒ†ã‚­ã‚¹ãƒˆã¨ãƒ•ã‚¡ã‚¤ãƒ«åï¼‰
text_input = widgets.Text(
    description='å†…å®¹:',
    placeholder='ã“ã“ã«URLã‚„ãƒ†ã‚­ã‚¹ãƒˆã‚’å…¥åŠ›',
    layout=widgets.Layout(width='80%')
)

filename_input = widgets.Text(
    description='ä¿å­˜å:',
    placeholder='ä¾‹: my_qrcode.png',
    layout=widgets.Layout(width='50%')
)

# ãƒœã‚¿ãƒ³
generate_button = widgets.Button(
    description='QRã‚³ãƒ¼ãƒ‰ç”Ÿæˆ',
    button_style='success'
)

save_button = widgets.Button(
    description='ä¿å­˜ã™ã‚‹',
    button_style='info'
)

# å‡ºåŠ›é ˜åŸŸ
output_area = widgets.Output()

# ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã«QRç”»åƒã‚’ä¿å­˜
qr_image = None

# QRã‚³ãƒ¼ãƒ‰ç”Ÿæˆå‡¦ç†
def on_generate_clicked(b):
    global qr_image
    with output_area:
        clear_output()
        data = text_input.value.strip()
        if not data:
            print("âš ï¸ å…¥åŠ›ãŒç©ºã§ã™ã€‚")
            return
        # QRã‚³ãƒ¼ãƒ‰ç”Ÿæˆ
        qr = qrcode.make(data)
        qr_image = qr
        print("âœ… QRã‚³ãƒ¼ãƒ‰ãŒç”Ÿæˆã•ã‚Œã¾ã—ãŸã€‚")
        display(qr_image)

# ä¿å­˜å‡¦ç†
def on_save_clicked(b):
    global qr_image
    with output_area:
        if qr_image is None:
            print("âš ï¸ QRã‚³ãƒ¼ãƒ‰ãŒã¾ã ç”Ÿæˆã•ã‚Œã¦ã„ã¾ã›ã‚“ã€‚")
            return
        filename = filename_input.value.strip()
        if not filename:
            print("âš ï¸ ä¿å­˜ãƒ•ã‚¡ã‚¤ãƒ«åã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ã€‚")
            return
        if not filename.lower().endswith('.png'):
            filename += '.png'
        qr_image.save(filename)
        print(f"âœ… QRã‚³ãƒ¼ãƒ‰ã‚’ã€Œ{filename}ã€ã¨ã—ã¦ä¿å­˜ã—ã¾ã—ãŸã€‚ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã‚’é–‹å§‹ã—ã¾ã™ã€‚")
        files.download(filename)

# ãƒœã‚¿ãƒ³ã‚¤ãƒ™ãƒ³ãƒˆã®ç™»éŒ²
generate_button.on_click(on_generate_clicked)
save_button.on_click(on_save_clicked)

# UIã®è¡¨ç¤º
display(widgets.VBox([
    text_input,
    widgets.HBox([generate_button, save_button]),
    filename_input,
    output_area
]))
```
