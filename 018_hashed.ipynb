### 🔐 ハッシュ化とは？
ハッシュ化（ハッシュ関数）は、「**あるデータを一定のルールに従って固定長の文字列に変換する処理**」のことです。  
例えば、名前やパスワード、文章など、どんな長さのデータでも、ハッシュ化すると「決まった長さの文字列」になります。

---

### 🛠 ハッシュ化ツールって？
これは、入力したデータ（例：パスワード）を **SHA256やMD5** という方式でハッシュ化してくれるツールです。  
つまり、簡単に言えば：

> 🔡 入力（例：「password123」）  
> ↓  
> 🔄 ハッシュ化ツールで変換  
> ↓  
> 🧱 出力（例：`ef92b778bafe771e89245b89ecbc08a44a4e166c06659911881f383d4473e94f` ←SHA256）

---

### 💡 よく使われるハッシュ方式

| 方式   | 出力の長さ | 特徴                         |
|--------|-------------|------------------------------|
| MD5    | 128ビット（32文字） | 速度は速いが安全性は低い（現在は非推奨） |
| SHA-256 | 256ビット（64文字） | 安全性が高く、現在主流の方式            |

---

### 🧰 使い道の例
- パスワード保存（元のパスワードは保存せず、ハッシュだけ保存）
- データの改ざんチェック（ファイルのハッシュと一致するかで確認）
- ブロックチェーン（Bitcoinなどでも使われる）

## 🔐 ハッシュ関数の基本的な仕組み・数理的特徴

### ✅ 1. **入力 → 出力：一方向関数（One-way function）**
ハッシュ関数 \( H(x) \) は以下のような**一方向性**を持ちます：

- \( x \) → \( H(x) \) は**簡単**に計算できる。
- しかし、\( H(x) \) → \( x \) を**逆算するのは非常に困難（事実上不可能）**。

これは、たとえるなら「卵をスクランブルエッグにするのは簡単だけど、元に戻すのは無理」みたいな感じです🥚🍳

---

### ✅ 2. **固定長出力**
どんな長さの入力データでも、**出力は一定のビット長**になります。

例：SHA-256 → 256ビット（64桁の16進数）  
例：MD5 → 128ビット（32桁の16進数）

数学的には、  
\[ H: \{0,1\}^* \rightarrow \{0,1\}^n \]  
（任意長のビット列 → 固定長 \( n \) ビット）

---

### ✅ 3. **衝突耐性（Collision Resistance）**
異なる入力 \( x \neq y \) に対して：

\[
H(x) \neq H(y)
\]

となるように設計されています。つまり、**異なるデータが同じハッシュにならない**ように工夫されています（ただし理論上は無限の入力→有限の出力なので完全には防げません。だから衝突が「見つかりにくい」ことが重要）。

---

### ✅ 4. **微小な変化が大きく変化を生む（Avalanche Effect）**
入力を1ビットだけ変えても、**ハッシュ値は全く異なるものになります**。

例：
```
H("hello") = 2cf24dba5fb0a30e...  
H("Hello") = 185f8db32271fe25...
```

これを「**アバランチ効果**」と呼びます。

---

## ⚙️ どんな数学が使われているの？

### 📘 1. **ビット演算（XOR, AND, OR, シフト）**
ハッシュ関数では、高速で計算できる**ビット単位の操作**が多用されます。これはCPUにとって効率がよく、簡単に並列化できるためです。

---

### 📘 2. **置換・転置（Permutations and Mixings）**
入力のビット列を、決められたパターンで**ぐちゃぐちゃに混ぜる処理**が何段階も行われます。これにより、アバランチ効果が生まれます。

---

### 📘 3. **論理ブロック構造（Merkle–Damgård構造など）**
多くのハッシュ関数（MD5、SHA-1、SHA-2など）は、以下の構造を使います：

1. 入力データを一定サイズのブロックに分ける（例：512ビット）
2. 初期ハッシュ値（IV）からスタート
3. 各ブロックに対して圧縮関数を繰り返し適用
4. 最後の出力がハッシュ値

これを「**Merkle–Damgård構造**」と呼びます。

---

### 📘 4. **置換ボックス（S-box）**
MD5やSHAシリーズなどでは、特定の入力に対して出力が一見ランダムになるように設計された「**S-box（Substitution box）**」を使用します。  
これは暗号化の世界でもよく使われる数学的構造で、**非線形性**を持たせてセキュリティを高めるために使われます。

---

### 📘 5. **数論・群論（特に新しい方式）**
SHA-3やBLAKE2など、より新しいハッシュ関数では、高度な代数的構造や有限体の演算なども登場します。  
たとえば SHA-3 は「**Keccak（ケッチャク）**」という構造を使っており、ビットの行列を変形・回転・XORなどで操作します。

## 🧠 まとめ：ハッシュ関数が満たすべき数学的性質

| 性質            | 説明 |
|-----------------|------|
| 衝突耐性         | 同じハッシュ値を持つ異なるデータを作るのが難しい |
| 一方向性         | ハッシュ値から元のデータを復元できない |

## 🔧 SHA-256のビット操作：ざっくり全体像

SHA-256 は、以下のような流れでハッシュ値を作ります：

1. **入力データの前処理（パディング）**
2. **512ビットずつブロック分割**
3. **各ブロックに対する64ラウンドの処理（メイン部）**
4. **最終的なハッシュ値の出力**

ここで「**64ラウンドの処理**」の中が、ビット演算でゴリゴリやってる部分です！  
では、その中身を見ていきましょう。

---

## ⚙️ 64ラウンドの内部処理：SHA-256の主なビット操作

以下のようなビット演算がメインで使われています：

### 🌀 1. **ビットシフト（右シフト、ローテーション）**

- **右シフト（Shift right）**: ビット列を右にずらす  
- **ローテーション（Rotate right）**: ビット列を右に回転させる（はみ出したビットは左端に戻る）

```text
例：
x = 10110010 （8ビット）
右シフト → 01011001
右ローテーション → 01011001（似てるが少し違う！）
```

SHA-256ではこんな風に使われます👇

```python
S1 = ROTR(e, 6) ^ ROTR(e, 11) ^ ROTR(e, 25)
```

（`ROTR(x, n)` は「x を n ビット右ローテート」）

---

### 🧮 2. **XOR（排他的論理和）**

```text
A XOR B
→ ビットが違えば1、同じなら0

例：
A = 1011
B = 1100
A XOR B = 0111
```

これは、情報を「混ぜてシャッフルする」目的でよく使われます。

---

### 🧲 3. **AND / OR**

- AND（論理積）：両方1なら1  
- OR（論理和）：どちらかが1なら1

SHA-256では次のような「選択関数」があります：

```text
Ch(x, y, z) = (x AND y) XOR ((NOT x) AND z)
```

これは「xに応じてyかzを選ぶ」処理になっていて、条件分岐的な使い方ができます。

---

### 🧱 4. **定数の加算（mod 2^32）**

SHA-256では、32ビットの定数（K）と各ステップでの計算値を足し合わせます。

```text
T1 = h + S1 + Ch(e, f, g) + K[i] + W[i]
```

ここでの加算は「**オーバーフローしてもいいよ**」という**modulo 2³²**で行われるため、ビット単位の操作になります。

---

## 📦 各ラウンドのビット操作まとめ（ざっくりフロー）

1. **e, f, g** などのレジスタに対してビットローテーション・XOR
2. **選択関数 `Ch(e, f, g)`** で分岐処理
3. **定数 `K[i]` や拡張ワード `W[i]` を加算**
4. **一時変数 T1, T2 を計算して状態レジスタを更新**
5. これを **64回繰り返す！**

---

## 🧠 実際のPython擬似コード（簡易）

```python
def ROTR(x, n):
    return (x >> n) | (x << (32 - n)) & 0xFFFFFFFF

def Ch(x, y, z):
    return (x & y) ^ (~x & z)

def Maj(x, y, z):
    return (x & y) ^ (x & z) ^ (y & z)

# 各ステップでの例（超簡略化）
S1 = ROTR(e, 6) ^ ROTR(e, 11) ^ ROTR(e, 25)
ch = Ch(e, f, g)
temp1 = h + S1 + ch + K[i] + W[i]
```

## 🔢 SHA-3（Keccak）の仕組み：行列と代数の世界

SHA-3 は従来の SHA-2 などと違って、**スポンジ構造（Sponge Construction）**という新しい構造を使っています。  
ビット列の「海綿（スポンジ）」にデータを吸い込ませて、ギュッと絞るイメージです🧽

### 特徴的な処理：
- **5×5×w のビット行列（状態行列）**：これは SHA-3 の「心臓部」
- 各ラウンドで次の操作を行います：
  1. θ（Theta）：列ごとの XOR 平均補正
  2. ρ（Rho）：ビットのローテーション
  3. π（Pi）：行列の再配置
  4. χ（Chi）：非線形変換（ビット単位で AND / NOT / XOR）
  5. ι（Iota）：ラウンド定数を加算

⚙️ → まるで「代数的パズル」みたいな仕組み！

## 🧪 Step 1: 最小限のセットアップ

まずは、Pythonで文字列をビット列に変換し、それを使って「簡単なハッシュ」を作る流れです👇

```python
def str_to_bits(s):
    return ''.join(f'{ord(c):08b}' for c in s)

def rotate_right(bits, n):
    return bits[-n:] + bits[:-n]

def simple_hash(message):
    # 1. 文字列 → ビット列
    bits = str_to_bits(message)
    
    # 2. 32ビットに切り詰め（またはパディング）
    bits = (bits + '0' * 32)[:32]
    
    # 3. 初期値（固定）
    h = '10101010101010101010101010101010'
    
    # 4. 8ラウンドの繰り返し（ローテーション＋XOR）
    for i in range(8):
        rotated = rotate_right(bits, i + 1)
        h = ''.join('1' if a != b else '0' for a, b in zip(h, rotated))
    
    return h

# 実行例
msg = "hello"
hash_value = simple_hash(msg)
print(f"Hash('{msg}') = {hash_value}")
```

次に進めるとしたら、例えば：

````python
def str_to_bits(s):
    return ''.join(f'{ord(c):08b}' for c in s)

def rotate_right(bits, n):
    return bits[-n:] + bits[:-n]

def simple_hash(message):
    # 1. 文字列 → ビット列
    bits = str_to_bits(message)

    # 2. 32ビットに切り詰め（またはパディング）
    bits = (bits + '0' * 32)[:32]

    # 3. 初期値（固定）
    h = '10101010101010101010101010101010'

    # 4. 8ラウンドの繰り返し（ローテーション＋XOR）
    for i in range(8):
        rotated = rotate_right(bits, i + 1)
        h = ''.join('1' if a != b else '0' for a, b in zip(h, rotated))

    return h

# 実行例
msg = "hello"
hash_value = simple_hash(msg)
print(f"Hash('{msg}') = {hash_value}")
````

気になる方向ある？次のステップ選んでもらえたら、そこを一緒に実装していこう！

````python
def str_to_bits(s):
    return ''.join(f'{ord(c):08b}' for c in s)

def rotate_right(bits, n):
    return bits[-n:] + bits[:-n]

def bin_to_hex(bin_str):
    return hex(int(bin_str, 2))[2:].zfill(len(bin_str) // 4)

# ラウンド定数（32ビットずつ）
ROUND_CONSTANTS = [
    '00110011001100110011001100110011',
    '01010101010101010101010101010101',
    '11110000111100001111000011110000',
    '00001111000011110000111100001111',
    '10101010101010101010101010101010',
    '11001100110011001100110011001100',
    '11111111000000001111111100000000',
    '00000000111111110000000011111111',
]

def simple_hash_v2(message):
    bits = str_to_bits(message)
    bits = (bits + '0' * 32)[:32]
    h = '10101010101010101010101010101010'

    for i in range(8):
        rotated = rotate_right(bits, i + 1)
        h = ''.join('1' if a != b else '0' for a, b in zip(h, rotated))
        h = ''.join('1' if a != b else '0' for a, b in zip(h, ROUND_CONSTANTS[i]))  # 定数XOR

    return bin_to_hex(h)

# 実行例
msg = "hello"
print(f"Hash('{msg}') = {simple_hash_v2(msg)}")
````

def bitwise_and(a, b):
    return ''.join('1' if x == '1' and y == '1' else '0' for x, y in zip(a, b))

def bitwise_or(a, b):
    return ''.join('1' if x == '1' or y == '1' else '0' for x, y in zip(a, b))

def bitwise_xor(a, b):
    return ''.join('1' if x != y else '0' for x, y in zip(a, b))

def bitwise_not(a):
    return ''.join('1' if x == '0' else '0' for x in a)

def Ch(x, y, z):
