```python
# @title ã‚µã‚¤ã‚³ãƒ­ã‚’"1å›"ã ã‘æŒ¯ã‚‹

import random
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display, clear_output
import ipywidgets as widgets
from collections import Counter
import io

# å‡ºç›®ã®å±¥æ­´ã‚’ä¿å­˜ã™ã‚‹ãƒªã‚¹ãƒˆ
roll_history = []

# ç”»åƒç”Ÿæˆï¼ˆæ•°å­—ã‚’æç”»ï¼‰
def generate_number_image(number):
    img = Image.new('RGB', (100, 100), color='white')
    draw = ImageDraw.Draw(img)

    try:
        font = ImageFont.truetype("DejaVuSans-Bold.ttf", 60)
    except:
        font = ImageFont.load_default()

    # textbboxã§æ–‡å­—ã‚µã‚¤ã‚ºã‚’å–å¾—
    bbox = draw.textbbox((0, 0), str(number), font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]

    # ä¸­å¤®ã«æç”»
    position = ((100 - text_width) / 2, (100 - text_height) / 2)
    draw.text(position, str(number), fill='black', font=font)

    return img

# ã‚°ãƒ©ãƒ•è¡¨ç¤º
def show_frequency_graph(history):
    counts = Counter(history)
    numbers = list(range(10))
    frequencies = [counts.get(i, 0) for i in numbers]

    plt.figure(figsize=(8, 4))
    plt.bar(numbers, frequencies, color='skyblue')
    plt.xlabel("æ•°å­—")
    plt.ylabel("å‡ºç¾å›æ•°")
    plt.title("å‡ºç›®ã®ç¢ºç‡ï¼ˆé »åº¦ï¼‰")
    plt.xticks(numbers)
    plt.grid(axis='y')
    plt.show()

# å‡ºåŠ›ã‚¨ãƒªã‚¢
output_area = widgets.Output()

# ãƒœã‚¿ãƒ³
roll_button = widgets.Button(
    description='ã‚µã‚¤ã‚³ãƒ­ã‚’æŒ¯ã‚‹',
    button_style='success',
    layout=widgets.Layout(width='200px')
)

# ãƒœã‚¿ãƒ³ãŒæŠ¼ã•ã‚ŒãŸæ™‚ã®å‹•ä½œ
def on_roll_clicked(b):
    with output_area:
        clear_output(wait=True)
        number = random.randint(0, 9)
        roll_history.append(number)

        # ç”»åƒè¡¨ç¤º
        img = generate_number_image(number)
        display(img)

        # ãƒ†ã‚­ã‚¹ãƒˆè¡¨ç¤º
        print(f"å‡ºãŸæ•°å­—ï¼š{number}")

        # ã‚°ãƒ©ãƒ•è¡¨ç¤º
        show_frequency_graph(roll_history)

roll_button.on_click(on_roll_clicked)

# è¡¨ç¤º
display(widgets.VBox([roll_button, output_area]))
```

```python
# @title ã‚µã‚¤ã‚³ãƒ­æ”¹è‰¯ç‰ˆ

import random
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display, clear_output
import ipywidgets as widgets
from collections import Counter

# å‡ºç›®å±¥æ­´ã‚’ä¿å­˜ã™ã‚‹ãƒªã‚¹ãƒˆ
roll_history = []

# -------------------------
# è¨­å®šï¼ˆãƒ¦ãƒ¼ã‚¶ãƒ¼èª¿æ•´å¯èƒ½ï¼‰
# -------------------------
FONT_SIZE = 60
FONT_COLOR = "blue"
IMAGE_WIDTH = 120
IMAGE_HEIGHT = 120

# ç”»åƒç”Ÿæˆï¼ˆæ•°å­—ã‚’æç”»ï¼‰
def generate_number_image(number):
    img = Image.new('RGB', (IMAGE_WIDTH, IMAGE_HEIGHT), color='white')
    draw = ImageDraw.Draw(img)

    try:
        font = ImageFont.truetype("DejaVuSans-Bold.ttf", FONT_SIZE)
    except:
        font = ImageFont.load_default()

    # ãƒ†ã‚­ã‚¹ãƒˆã®ãƒã‚¦ãƒ³ãƒ‡ã‚£ãƒ³ã‚°ãƒœãƒƒã‚¯ã‚¹ã‚’å–å¾—
    bbox = draw.textbbox((0, 0), str(number), font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]

    # ä¸­å¤®ã«æç”»
    position = ((IMAGE_WIDTH - text_width) / 2, (IMAGE_HEIGHT - text_height) / 2)
    draw.text(position, str(number), fill=FONT_COLOR, font=font)

    return img

# å‡ºç›®é »åº¦ã‚°ãƒ©ãƒ•è¡¨ç¤º
def show_frequency_graph(history):
    counts = Counter(history)
    numbers = list(range(10))
    frequencies = [counts.get(i, 0) for i in numbers]

    plt.figure(figsize=(8, 4))
    plt.bar(numbers, frequencies, color='skyblue')
    plt.xlabel("æ•°å­—")
    plt.ylabel("å‡ºç¾å›æ•°")
    plt.title("å‡ºç›®ã®ç¢ºç‡ï¼ˆé »åº¦ï¼‰")
    plt.xticks(numbers)
    plt.grid(axis='y')
    plt.show()

# å‡ºåŠ›ã‚¨ãƒªã‚¢
output_area = widgets.Output()

# ã€Œã‚µã‚¤ã‚³ãƒ­ã‚’æŒ¯ã‚‹ã€ãƒœã‚¿ãƒ³
roll_button = widgets.Button(
    description='ã‚µã‚¤ã‚³ãƒ­ã‚’æŒ¯ã‚‹',
    button_style='success',
    layout=widgets.Layout(width='200px')
)

# ã€Œãƒªã‚»ãƒƒãƒˆã€ãƒœã‚¿ãƒ³
reset_button = widgets.Button(
    description='ãƒªã‚»ãƒƒãƒˆ',
    button_style='danger',
    layout=widgets.Layout(width='200px')
)

# ã‚µã‚¤ã‚³ãƒ­ã‚’æŒ¯ã‚‹ãƒœã‚¿ãƒ³ã®å‹•ä½œ
def on_roll_clicked(b):
    with output_area:
        clear_output(wait=True)
        number = random.randint(0, 9)
        roll_history.append(number)

        # ç”»åƒè¡¨ç¤º
        img = generate_number_image(number)
        display(img)

        # ãƒ†ã‚­ã‚¹ãƒˆè¡¨ç¤º
        print(f"å‡ºãŸæ•°å­—ï¼š{number}")

        # ã‚°ãƒ©ãƒ•è¡¨ç¤º
        show_frequency_graph(roll_history)

# ãƒªã‚»ãƒƒãƒˆãƒœã‚¿ãƒ³ã®å‹•ä½œ
def on_reset_clicked(b):
    with output_area:
        clear_output(wait=True)
        roll_history.clear()
        print("å‡ºç›®å±¥æ­´ã‚’ãƒªã‚»ãƒƒãƒˆã—ã¾ã—ãŸã€‚")

# ã‚¤ãƒ™ãƒ³ãƒˆç™»éŒ²
roll_button.on_click(on_roll_clicked)
reset_button.on_click(on_reset_clicked)

# è¡¨ç¤º
display(widgets.VBox([
    widgets.HBox([roll_button, reset_button]),
    output_area
]))
```

```python
# @title ã‚µã‚¤ã‚³ãƒ­nå›è‡ªå‹•å®Ÿè¡Œç‰ˆ

n = 10000000

import random
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display, clear_output
import ipywidgets as widgets
from collections import Counter

# å‡ºç›®å±¥æ­´ã‚’ä¿å­˜ã™ã‚‹ãƒªã‚¹ãƒˆ
roll_history = []

# å‡ºç›®é »åº¦ã‚°ãƒ©ãƒ•è¡¨ç¤º
def show_frequency_graph(history):
    counts = Counter(history)
    numbers = list(range(10))
    frequencies = [counts.get(i, 0) for i in numbers]

    plt.figure(figsize=(8, 4))
    plt.bar(numbers, frequencies, color='pink')
    plt.xlabel("number")
    plt.ylabel("occurrences")
    plt.title("probability")
    plt.xticks(numbers)
    plt.grid(axis='y')
    plt.show()

# å‡ºåŠ›ã‚¨ãƒªã‚¢
output_area = widgets.Output()

# ã‚«ã‚¦ãƒ³ãƒˆåˆæœŸåŒ–
counts = 0

# ã‚µã‚¤ã‚³ãƒ­ã‚’100å›æŒ¯ã‚‹
with output_area:
    clear_output(wait=True)
    print(f"ã‚µã‚¤ã‚³ãƒ­ã‚’{n}å›è‡ªå‹•ã§æŒ¯ã‚Šã¾ã™...\n")
    while counts < n:
        number = random.randint(0, 9)
        roll_history.append(number)
        counts += 1

    # å‡ºç›®ä¸€è¦§ã®è¡¨ç¤º
    print("å‡ºãŸæ•°å­—ã®ä¸€è¦§ï¼ˆå…ˆé ­20å€‹ã®ã¿è¡¨ç¤ºï¼‰ï¼š")
    print(roll_history[:20], "...")
    print(f"\nåˆè¨ˆï¼š{len(roll_history)}å›")

    # ã‚°ãƒ©ãƒ•è¡¨ç¤º
    show_frequency_graph(roll_history)

# è¡¨ç¤º
display(output_area)
```
