```python
# ステップ 0: 練習用CSVファイルの生成（実データがない場合）

import pandas as pd
import numpy as np

# 練習用のデータフレームを生成
np.random.seed(0)
sample_data = pd.DataFrame({
    'ID': range(1, 101),
    '年齢': np.random.randint(18, 65, 100),
    '身長(cm)': np.random.normal(165, 10, 100).round(1),
    '体重(kg)': np.random.normal(60, 12, 100).round(1),
    '性別': np.random.choice(['男性', '女性'], 100),
    '部署': np.random.choice(['営業', '開発', '人事', '総務'], 100)
})

# CSVファイルとして保存
sample_csv_name = 'sample_data.csv'
sample_data.to_csv(sample_csv_name, index=False)
print(f"練習用CSV「{sample_csv_name}」を生成しました。")

# 読み込み
df = pd.read_csv(sample_csv_name)
df.head()
```

```python
# 必要なライブラリをインポート
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# グラフのスタイル設定（任意）
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# CSVファイルのアップロード
# uploaded = files.upload()

# 最初のファイル名を取得して読み込む（自動的に1つ目のファイルを選択）
# filename = list(uploaded.keys())[0]
# try:
#     df = pd.read_csv(filename)
# except UnicodeDecodeError:
#     df = pd.read_csv(filename, encoding='shift_jis')

# データの冒頭を表示
df.head()

print("データの形状（行, 列）:", df.shape)
print("\nカラム名一覧:\n", df.columns.tolist())
print("\nデータ型:\n", df.dtypes)
print("\n欠損値のカウント:\n", df.isnull().sum())

# 基本統計量の表示（数値型に限定）
df.describe()

# 数値型カラムを自動取得
numeric_columns = df.select_dtypes(include='number').columns.tolist()

# 任意のカラムをヒストグラムで表示（例：最初の数値カラム）
if numeric_columns:
    col = numeric_columns[0]
    df[col].hist(bins=20)
    plt.title(f'ヒストグラム: {col}')
    plt.xlabel(col)
    plt.ylabel('頻度')
    plt.show()
else:
    print("数値カラムが見つかりません。")

# 数値型が2つ以上あるときのみ実行
if len(numeric_columns) >= 2:
    x_col = numeric_columns[0]
    y_col = numeric_columns[1]
    sns.scatterplot(data=df, x=x_col, y=y_col)
    plt.title(f'散布図: {x_col} vs {y_col}')
    plt.show()

# 相関行列のヒートマップ
if len(numeric_columns) >= 2:
    corr = df[numeric_columns].corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title('相関係数ヒートマップ')
    plt.show()

```

# データ分析ツール 仕様書（Google Colab用）

## 1. ツール概要

本ツールは、PandasとMatplotlibを用いてCSV等の表形式データを読み込み、基本的な統計分析および可視化を行うGoogle Colab上のPythonアプリケーションです。主に初心者から中級者向けのデータ探索・可視化の支援を目的としています。

---

## 2. 使用技術

* 言語: Python 3.x
* 実行環境: Google Colaboratory
* ライブラリ:

  * `pandas`: データの読み込みと処理
  * `matplotlib`: データの可視化
  * `seaborn`（任意）: より洗練された可視化（拡張オプション）

---

## 3. 主な機能

### 3.1 データ読み込み機能

* 対応ファイル形式: CSV（UTF-8 / Shift\_JIS 対応）
* Google Colabのファイルアップローダーを使用
* 読み込んだデータの先頭5行を表示（`df.head()`）

### 3.2 データ基本情報の表示

* 行数・列数の表示（`df.shape`）
* カラム名一覧（`df.columns`）
* 欠損値の有無とカウント（`df.isnull().sum()`）
* データ型の確認（`df.dtypes`）

### 3.3 基本統計量の表示

* 平均、中央値、最大値、最小値、標準偏差など（`df.describe()`）

* 棒グラフ：カテゴリカル変数の頻度表示

### 3.5 オプション機能（拡張予定）
